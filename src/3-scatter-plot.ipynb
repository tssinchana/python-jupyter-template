### Exercise 3: Scatter Plot with Annotations

```python
# TASK: Create a scatter plot with x = [1, 2, 3, 4, 5] and y = [2, 4, 6, 8, 10].
# Annotate each point with its (x, y) value, and set the title as 'Scatter Plot with Annotations'.
```

```python
import matplotlib.pyplot as plt

def validate_data(x, y):
    if not isinstance(x, list) or not isinstance(y, list):
        raise ValueError("x and y must be lists.")
    if len(x) == 0 or len(y) == 0:
        raise ValueError("x and y lists cannot be empty.")
    if len(x) != len(y):
        raise ValueError("x and y lists must have the same length.")
    if any(not isinstance(value, (int, float)) for value in x):
        raise ValueError("All values in x must be numbers.")
    if any(not isinstance(value, (int, float)) for value in y):
        raise ValueError("All values in y must be numbers.")

# Data
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

try:
    # Validate data
    validate_data(x, y)

    # Create scatter plot
    plt.scatter(x, y, color='b')

    # Add title and labels
    plt.title("Scatter Plot with Annotations")
    plt.xlabel("X")
    plt.ylabel("Y")

    # Add annotations for each point
    for i in range(len(x)):
        plt.annotate(f'({x[i]}, {y[i]})', (x[i], y[i]), textcoords="offset points", xytext=(0,10), ha='center')

    # Add grid lines
    plt.grid(True)

    # Show plot
    plt.show()

except ValueError as e:
    print(f"Data validation error: {e}")
except Exception as e:
    print(f"An unexpected error occurred: {e}")
```
