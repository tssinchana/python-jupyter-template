### Exercise 6: Histogram Customization

```python
# TASK: Create a histogram for the following data: data = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4, 5].
# Customize the histogram with a title, labels for the x-axis, and a specific color for the bars.
```

```python
import matplotlib.pyplot as plt

def validate_data(data):
    if not isinstance(data, list):
        raise ValueError("Data must be a list.")
    if len(data) == 0:
        raise ValueError("Data list cannot be empty.")
    if any(not isinstance(value, (int, float)) for value in data):
        raise ValueError("All values in data must be numbers.")

# Data
data = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4, 5]

try:
    # Validate data
    validate_data(data)

    # Create histogram
    plt.hist(data, bins=5, color='purple', edgecolor='black')
    plt.title("Histogram")
    plt.xlabel("Values")
    plt.ylabel("Frequency")
    
    # Show plot
    plt.show()

except ValueError as e:
    print(f"Data validation error: {e}")
except Exception as e:
    print(f"An unexpected error occurred: {e}")
```
