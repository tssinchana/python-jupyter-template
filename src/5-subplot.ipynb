### Exercise 5: Subplot Layout

```python
# TASK: Create a 2x2 subplot layout.
# Plot a line chart in the first subplot, a bar chart in the second, a scatter plot in the third, and a pie chart in the fourth.
```

```python
import matplotlib.pyplot as plt

def validate_data(x, y, categories, values, sizes):
    if not isinstance(x, list) or not isinstance(y, list):
        raise ValueError("x and y must be lists.")
    if len(x) == 0 or len(y) == 0:
        raise ValueError("x and y lists cannot be empty.")
    if len(x) != len(y):
        raise ValueError("x and y lists must have the same length.")
    if any(not isinstance(value, (int, float)) for value in x):
        raise ValueError("All values in x must be numbers.")
    if any(not isinstance(value, (int, float)) for value in y):
        raise ValueError("All values in y must be numbers.")
    if not isinstance(categories, list) or not isinstance(values, list):
        raise ValueError("Categories and values must be lists.")
    if len(categories) == 0 or len(values) == 0:
        raise ValueError("Categories and values lists cannot be empty.")
    if len(categories) != len(values):
        raise ValueError("Categories and values lists must have the same length.")
    if any(not isinstance(category, str) for category in categories):
        raise ValueError("All categories must be strings.")
    if any(not isinstance(value, (int, float)) for value in values):
        raise ValueError("All values must be numbers.")
    if not isinstance(sizes, list):
        raise ValueError("Sizes must be a list.")
    if len(sizes) == 0:
        raise ValueError("Sizes list cannot be empty.")
    if any(not isinstance(size, (int, float)) for size in sizes):
        raise ValueError("All sizes must be numbers.")

x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 9]
sizes = [40, 25, 20, 15]

try:
    # Validate data
    validate_data(x, y, categories, values, sizes)
    
    fig, axs = plt.subplots(2, 2, figsize=(10, 10))

    # First subplot: Line plot
    axs[0, 0].plot(x, y)
    axs[0, 0].set_title('Line Plot')

    # Second subplot: Bar plot
    axs[0, 1].bar(categories, values)
    axs[0, 1].set_title('Bar Plot')

    # Third subplot: Scatter plot
    axs[1, 0].scatter(x, y)
    axs[1, 0].set_title('Scatter Plot')

    # Fourth subplot: Pie chart
    axs[1, 1].pie(sizes, labels=categories, autopct='%1.1f%%')
    axs[1, 1].set_title('Pie Chart')

    plt.tight_layout()
    plt.show()

except ValueError as e:
    print(f"Data validation error: {e}")
except Exception as e:
    print(f"An unexpected error occurred: {e}")
```
